Just some code to re-plot the Matlab plots for the molecule MOT simulation.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Plot results of varying laser beam waist.

```python
timestamps = ['20240911_0927', '20240910_2054', '20240911_0050']
waists = ["5 mm", "7 mm", "9 mm"]

#acceleration vs position

for i in range(len(timestamps)):
    df = pd.read_csv("AccelVsPos" + timestamps[i] + ".csv", header=None)
    pos = df.iloc[:, 0].to_numpy()
    accel = df.iloc[:, 1].to_numpy()
    plt.plot(pos, accel, label=waists[i])

plt.xlabel('Position $d$ (mm)')
plt.ylabel('Acceleration $a_d$ (mm/ms$^2$)')
plt.xlim(-15.0, 15.0)
plt.ylim(-2.5, 2.5)
plt.grid(linestyle = '--', which = 'major', linewidth = 1, color='black')
plt.grid(linestyle = ':', which = 'minor', linewidth = 0.5, color = 'gray')
plt.minorticks_on()
#plt.savefig('./VaryBeamWaist/AccelVsPos.png')
plt.title('Accel. vs Pos. in SrF Capture MOT, no vibrational branching')
plt.legend(title='Laser beam waist')
plt.show()

#acceleration vs velocity
for i in range(len(timestamps)):
    df = pd.read_csv("AccelVsVel" + timestamps[i] + ".csv", header=None)
    vel = df.iloc[:, 0].to_numpy()
    accel = df.iloc[:, 1].to_numpy()
    plt.plot(vel, accel, label=waists[i])

plt.xlabel('Velocity $v_d$ (m/s)')
plt.ylabel('Acceleration $a_d$ (mm/ms$^2$)')
plt.xlim(-20.0, 20.0)
plt.ylim(-8.0, 8.0)
plt.title('Accel. vs Vel. in SrF Capture MOT, no vibrational branching')
plt.grid(linestyle = '--', which = 'major', linewidth = 1, color='black')
plt.grid(linestyle = ':', which = 'minor', linewidth = 0.5, color = 'gray')
plt.minorticks_on()
#plt.savefig('./VaryBeamWaist/AccelVsVel.png')
plt.legend(title='Laser beam waist')
plt.show()


#MOT capture trajectory
for i in range(len(timestamps)):
    df = pd.read_csv("CaptureVelocityTrajectory" + timestamps[i] + ".csv", header=None)
    pos = df.iloc[:, 0].to_numpy()
    vel = df.iloc[:, 1].to_numpy()
    plt.plot(pos, vel, label=waists[i])

plt.xlabel('d (mm)')
plt.ylabel('v$_d$ (m/s)')
plt.xlim(-15, 15)
plt.ylim(-6, 16)
plt.title('Molecule Trajectory into Capture MOT')
plt.grid(linestyle = '--', which = 'major', linewidth = 1, color='black')
plt.grid(linestyle = ':', which = 'minor', linewidth = 0.5, color = 'gray')
plt.minorticks_on()
#plt.savefig('./VaryBeamWaist/MolTrajIntoMOT.png')
plt.legend(title='Laser beam waist')
plt.show()


#in-MOT trajectory, no photon scatter
for i in range(len(timestamps)):
    df = pd.read_csv("MOTTrajectoryNoScatter" + timestamps[i] + ".csv", header=None)
    pos = df.iloc[:, 0].to_numpy()
    vel = df.iloc[:, 1].to_numpy()
    plt.plot(pos, vel, label=waists[i])

plt.xlabel('d (mm)')
plt.ylabel('v$_d$ (m/s)')
plt.xlim(-10, 10)
plt.ylim(-5, 5)
plt.title('Trajectory of Captured Molecule in MOT, no spontaneous scatter')
plt.grid(linestyle = '--', which = 'major', linewidth = 1, color='black')
plt.grid(linestyle = ':', which = 'minor', linewidth = 0.5, color = 'gray')
plt.minorticks_on()
#plt.savefig('./VaryBeamWaist/MolTrajDuringMOT.png')
plt.legend(title='Laser beam waist')
plt.show()
```

```python
#in-MOT trajectory, yes photon scatter
for i in range(len(timestamps)):
    df = pd.read_csv("MOTTrajectoryYesScatter" + timestamps[i] + ".csv", header=None)
    pos = df.iloc[:, 0].to_numpy()
    vel = df.iloc[:, 1].to_numpy()
    plt.plot(pos, vel, label=waists[i])

plt.xlabel('d (mm)')
plt.ylabel('v$_d$ (m/s)')
plt.title('Trajectory of Captured Molecule in MOT, with spontaneous scatter')
plt.legend(title='Laser beam waist')
plt.show()
```

Plot results of varying B-field gradient

```python
timestamps = ['20240911_1333', '20240910_2054', '20240911_1804']
gradients = ["5 G/cm", "8 G/cm", "15 G/cm"]

#acceleration vs position

for i in range(len(timestamps)):
    df = pd.read_csv("AccelVsPos" + timestamps[i] + ".csv", header=None)
    pos = df.iloc[:, 0].to_numpy()
    accel = df.iloc[:, 1].to_numpy()
    plt.plot(pos, accel, label=gradients[i])

plt.xlabel('Position (mm)')
plt.ylabel('Acceleration (mm/ms$^2$)')
plt.title('Acceleration vs Position')
plt.legend(title='B-field gradient')
plt.show()

#acceleration vs velocity
for i in range(len(timestamps)):
    df = pd.read_csv("AccelVsVel" + timestamps[i] + ".csv", header=None)
    vel = df.iloc[:, 0].to_numpy()
    accel = df.iloc[:, 1].to_numpy()
    plt.plot(vel, accel, label=gradients[i])

plt.xlabel('Velocity (m/s)')
plt.ylabel('Acceleration (mm/ms$^2$)')
plt.title('Acceleration vs Velocity')
plt.legend(title='B-field gradient')
plt.show()


#MOT capture trajectory
for i in range(len(timestamps)):
    df = pd.read_csv("CaptureVelocityTrajectory" + timestamps[i] + ".csv", header=None)
    pos = df.iloc[:, 0].to_numpy()
    vel = df.iloc[:, 1].to_numpy()
    plt.plot(pos, vel, label=gradients[i])

plt.xlabel('d (mm)')
plt.ylabel('v$_d$ (m/s)')
plt.title('Molecule Trajectory into Capture MOT')
plt.legend(title='B-field gradient')
plt.show()


#in-MOT trajectory
for i in range(len(timestamps)):
    df = pd.read_csv("MOTTrajectoryNoScatter" + timestamps[i] + ".csv", header=None)
    pos = df.iloc[:, 0].to_numpy()
    vel = df.iloc[:, 1].to_numpy()
    plt.plot(pos, vel, label=gradients[i])

plt.xlabel('d (mm)')
plt.ylabel('v$_d$ (m/s)')
plt.title('Trajectory of Captured Molecule in MOT, no spontaneous scatter')
plt.legend(title='B-field gradient')
plt.show()
```
