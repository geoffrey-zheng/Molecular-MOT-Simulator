Just some code to re-plot the Matlab plots for the molecule MOT simulation.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Plot results of varying laser beam waist.

```python
timestamps = ['20240911_0927', '20240910_2054', '20240911_0050']
waists = ["5 mm", "7 mm", "9 mm"]

#acceleration vs position

for i in range(len(timestamps)):
    df = pd.read_csv("AccelVsPos" + timestamps[i] + ".csv", header=None)
    pos = df.iloc[:, 0].to_numpy()
    accel = df.iloc[:, 1].to_numpy()
    plt.plot(pos, accel, label=waists[i], linewidth=2.0)

plt.xlabel('Position $d$ (mm)')
plt.ylabel('Acceleration $a_d$ (mm/ms$^2$)')
plt.xlim(-15.0, 15.0)
plt.ylim(-2.5, 2.5)

ax = plt.gca()  # Get the current Axes instance
# Add additional spines in the middle
ax.spines['middle_x'] = plt.Line2D((0.5, 0.5), (0, 1), transform=ax.transAxes, color='black', linewidth=2)
ax.spines['middle_y'] = plt.Line2D((0, 1), (0.5, 0.5), transform=ax.transAxes, color='black', linewidth=2)

# Add the new spines to the axes
ax.add_line(ax.spines['middle_x'])
ax.add_line(ax.spines['middle_y'])

plt.grid(linestyle = '--', which = 'major', linewidth = 1, color='black')
plt.grid(linestyle = ':', which = 'minor', linewidth = 0.5, color = 'gray')
plt.minorticks_on()
plt.title('Accel. vs Pos. in SrF Capture MOT, no vibrational branching')
plt.legend(title='Laser beam waist')
#plt.savefig('./VaryBeamWaist/AccelVsPos.png')
plt.show()


#acceleration vs velocity
for i in range(len(timestamps)):
    df = pd.read_csv("AccelVsVel" + timestamps[i] + ".csv", header=None)
    vel = df.iloc[:, 0].to_numpy()
    accel = df.iloc[:, 1].to_numpy()
    plt.plot(vel, accel, label=waists[i], linewidth=2.0)

plt.xlabel('Velocity $v_d$ (m/s)')
plt.ylabel('Acceleration $a_d$ (mm/ms$^2$)')
plt.xlim(-20.0, 20.0)
plt.ylim(-8.0, 8.0)

ax = plt.gca()  # Get the current Axes instance
# Add additional spines in the middle
ax.spines['middle_x'] = plt.Line2D((0.5, 0.5), (0, 1), transform=ax.transAxes, color='black', linewidth=2)
ax.spines['middle_y'] = plt.Line2D((0, 1), (0.5, 0.5), transform=ax.transAxes, color='black', linewidth=2)

# Add the new spines to the axes
ax.add_line(ax.spines['middle_x'])
ax.add_line(ax.spines['middle_y'])

plt.title('Accel. vs Vel. in SrF Capture MOT, no vibrational branching')
plt.grid(linestyle = '--', which = 'major', linewidth = 1, color='black')
plt.grid(linestyle = ':', which = 'minor', linewidth = 0.5, color = 'gray')
plt.minorticks_on()
plt.legend(title='Laser beam waist')
#plt.savefig('./VaryBeamWaist/AccelVsVel.png')
plt.show()


#MOT capture trajectory
for i in range(len(timestamps)):
    df = pd.read_csv("CaptureVelocityTrajectory" + timestamps[i] + ".csv", header=None)
    pos = df.iloc[:, 0].to_numpy()
    vel = df.iloc[:, 1].to_numpy()
    plt.plot(pos, vel, label=waists[i], linewidth=2.0)

plt.xlabel('Position $d$ (mm)')
plt.ylabel('Velocity $v_d$ (m/s)')
plt.xlim(-15, 15)
plt.ylim(-5, 18)

ax = plt.gca()  # Get the current Axes instance
# Add additional spines in the middle
ax.spines['middle_x'] = plt.Line2D((0.5, 0.5), (0, 1), transform=ax.transAxes, color='black', linewidth=2)
ax.spines['middle_y'] = plt.Line2D((0, 1), (5/23, 5/23), transform=ax.transAxes, color='black', linewidth=2)

# Add the new spines to the axes
ax.add_line(ax.spines['middle_x'])
ax.add_line(ax.spines['middle_y'])

plt.title('SrF Molecule Trajectory into Capture MOT')
plt.grid(linestyle = '--', which = 'major', linewidth = 1, color='black')
plt.grid(linestyle = ':', which = 'minor', linewidth = 0.5, color = 'gray')
plt.minorticks_on()
plt.legend(title='Laser beam waist')
#plt.savefig('./VaryBeamWaist/MolTrajIntoMOT.png')
plt.show()



#in-MOT trajectory, no photon scatter
for i in range(len(timestamps)):
    df = pd.read_csv("MOTTrajectoryNoScatter" + timestamps[i] + ".csv", header=None)
    pos = df.iloc[:, 0].to_numpy()
    vel = df.iloc[:, 1].to_numpy()
    plt.plot(pos, vel, label=waists[i], linewidth=2.0)

plt.xlabel('Position $d$ (mm)')
plt.ylabel('Velocity $v_d$ (m/s)')
plt.xlim(-10, 10)
plt.ylim(-5, 5)

ax = plt.gca()  # Get the current Axes instance
# Add additional spines in the middle
ax.spines['middle_x'] = plt.Line2D((0.5, 0.5), (0, 1), transform=ax.transAxes, color='black', linewidth=2)
ax.spines['middle_y'] = plt.Line2D((0, 1), (0.5, 0.5), transform=ax.transAxes, color='black', linewidth=2)

# Add the new spines to the axes
ax.add_line(ax.spines['middle_x'])
ax.add_line(ax.spines['middle_y'])

plt.title('Trajectory of Captured SrF Molecule in MOT, no spontaneous scatter')
plt.grid(linestyle = '--', which = 'major', linewidth = 1, color='black')
plt.grid(linestyle = ':', which = 'minor', linewidth = 0.5, color = 'gray')
plt.minorticks_on()
plt.legend(title='Laser beam waist')
plt.savefig('./VaryBeamWaist/MolTrajDuringMOT.png')
plt.show()

```
